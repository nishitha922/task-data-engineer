```python
#Import all the necassary libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('/content/Melbourne_housing_FULL.csv')
df.head()
```

```python
#To check the column names
df.columns
```

```python
# To Identify rows with missing values
missing_rows = df[df.isnull().any(axis=1)]
```

```python
#Count the missing values per column
df.isnull().sum()
```

```python
print("Rows with missing values:")
missing_rows.head()
```

```python
# To identify any duplicate rows in the dataset
print("The number of duplicate entired:",df.duplicated().sum())
print("Duplicated record:",)
df[df.duplicated()]
```

```python
df.plot.box(figsize=(15,5))
```

```python
#Removal of outlier
upper_limit1=df['Price'].quantile(0.90)
lower_limit1=df['Price'].quantile(0.10)
df['Price']=np.where(df['Price']>=upper_limit1,upper_limit1,df['Price'])
df['Price']=np.where(df['Price']<=lower_limit1,lower_limit1,df['Price'])
```

```python
sns.boxplot(df['Price'])
```
